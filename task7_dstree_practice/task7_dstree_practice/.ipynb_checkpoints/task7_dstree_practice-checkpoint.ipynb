#  I. Немного решающих деревьев
##### Задача 1. Построение "среднего" алгоритма $\mathbb{E}_X [\mu(X)]$
В этом задании вам нужно построить графики, демонстрирующие, как алгоритм $\mathbb{E}_X [\mu(X)]$ аппроксимирует истинную зависимость в данных и как он меняется в зависимости от гиперпараметров метода обучения.

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
def f(x):
    return np.sin(x)   # истинная зависимость в данных
sample_size = 100      # длина выборки
samples_num = 20       # количество выборок
linspace = np.linspace(0, 7, 1000)  # точки для построения графиков
```

__1. (1 балл)__
1. Сгенерируйте выборку $x$ из одномерного экспоненциального распределения (np.random.exponential) длины sample_size.
1. Создайте вектор целевых переменных $y$ как сумму $f(x)$ и случайного шума, сгенерированного из равномерного распределения на отрезке $[-1, 1]$ (np.random.uniform).
1. Обучите DecisionTreeRegressor с параметрами по умолчанию на полученной выборке и сделайте предсказания для объектов из linspace.
1. Постройте два графика на одном рисунке: $f(x)$ и зависимость, восстановленную решающим деревом.

    Рекомендация: не забудьте, что все методы обучения в sklearn требуют на вход двумерную матрицу объекты-признаки. Сделать такую из одномерного вектора можно добавлением мнимых осей (np.newaxis).

```python
np.random.seed(42)
x = np.random.exponential(scale=1.0, size=sample_size)

y = f(x) + np.random.uniform(-1, 1, size=sample_size)

X_train = x.reshape(-1, 1) 
tree = DecisionTreeRegressor()
tree.fit(X_train, y)

X_test = linspace.reshape(-1, 1)
y_pred = tree.predict(X_test)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, label='Данные с шумом')
plt.plot(linspace, f(linspace), 'r-', linewidth=2, label='Истинная зависимость f(x)')
plt.plot(linspace, y_pred, 'g-', linewidth=2, label='Предсказание дерева')
plt.xlabel('x')
plt.ylabel('y')
plt.title('DecisionTreeRegressor на одной выборке')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

__2. (0.5 балла)__

Повторите первые 3 шага, описанные выше, samples_num раз. На одном графике для каждого обученного решающего дерева визуализируйте восстановленную им зависимость (рекомендуется все такие линии рисовать полупрозрачными и серым цветом: plt.plot(...... color="gray", alpha=0.5)).  На этом же графике изобразите истинную зависимость f(x) (красным цветом: color="red") и усредненную по всем деревьям восстановленную зависимость (черным цветом: color="black").    

```python
def depth(max_depth=None):
    plt.figure(figsize=(12, 8))
    
    all_predictions = []
    
    for i in range(samples_num):
        x = np.random.exponential(scale=1.0, size=sample_size)
        y = f(x) + np.random.uniform(-1, 1, size=sample_size)
        
        X_train = x[:, np.newaxis]
        tree = DecisionTreeRegressor(max_depth=max_depth)
        tree.fit(X_train, y)
        
        X_test = linspace[:, np.newaxis]
        y_pred = tree.predict(X_test)
        all_predictions.append(y_pred)
        
        plt.plot(linspace, y_pred, color='gray', alpha=0.3, linewidth=0.5)

    avg_prediction = np.mean(all_predictions, axis=0)

    plt.plot(linspace, f(linspace), 'r-', linewidth=3, label='Истинная зависимость')
    

    plt.plot(linspace, avg_prediction, 'k-', linewidth=3, label='Усредненное предсказание')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'DecisionTreeRegressor (max_depth={max_depth}) - {samples_num} выборок')
    plt.legend()
    plt.grid(True)
    plt.show()

depth(None)
```

Необходимо в каждую ячейку датафреймов bias и variance записать записать соответственно смещение и разброс, рассчитанные для алгоритма, задаваемого строкой таблицы, и генератора случайных чисел, задаваемого столбцом.
```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressors = [LinearRegression(), DecisionTreeRegressor(), RandomForestRegressor(), GradientBoostingRegressor()]

generators = [lambda size: np.random.normal(0, 0.3, size=size), 
              lambda size: np.random.exponential(0.3, size=size), 
              lambda size: np.random.uniform(-1, 1, size=size),
              lambda size: np.random.uniform(0, 1, size=size)]
```

```python
df_bias = pd.DataFrame(columns=["normal(0, 0.3)", "exponential(0.3)", "uniform(-1, 1)", "uniform(0, 1)"],
                       index=["LinearRegression", "DecisionTree", "RandomForest", "GradientBoosting"])
df_variance = pd.DataFrame(columns=df_bias.columns, index=df_bias.index)
```

```python
def f2(x):
    return 5 * np.sin(10*x)
```

```python
from sklearn.model_selection import train_test_split

def compute_bias_variance(estimator, sample_generator, n_samples=100, sample_size=100, 
                          n_objects=1000, f_true=f2, noise_generator=None):
    
    if noise_generator is None:
        noise_generator = sample_generator
    
    X_test = sample_generator(n_objects)[:, np.newaxis]
    y_true = f_true(X_test[:, 0])
    
    predictions = []
    
    for _ in range(n_samples):
        X_train = sample_generator(sample_size)[:, np.newaxis]
        y_train = f_true(X_train[:, 0]) + noise_generator(sample_size)
   
        estimator.fit(X_train, y_train)
        
        y_pred = estimator.predict(X_test)
        predictions.append(y_pred)
    
    predictions = np.array(predictions) 
    
    mean_prediction = np.mean(predictions, axis=0)
    
    bias_sq = np.mean((mean_prediction - y_true) ** 2)
    
    variance = np.mean(np.var(predictions, axis=0))
    
    return bias_sq, variance

for i, regressor in enumerate(regressors):
    for j, generator in enumerate(generators):
        bias, variance = compute_bias_variance(
            estimator=regressor,
            sample_generator=generator,
            n_samples=50,
            sample_size=200,
            n_objects=1000
        )
        
        df_bias.iloc[i, j] = bias
        df_variance.iloc[i, j] = variance

print("Смещение (Bias^2):")
print(df_bias)
print("\nРазброс (Variance):")
print(df_variance)
```

__7. (2 балла)__

Постройте графики зависимости смещения и разброса от sample_size, samples_num и objects_num (при варьировании одного параметра остальные два остаются по умолчанию) для четырех рассмотренных регрессионных алгоритмов. Сделайте выводы о том, как эти параметры влияют на дисперсию оценок.

Будем считать для нормального распределения

```python
def plot_parameter_vs_bias_variance(parameter_name, parameter_values, default_params):
    """
    Строит графики зависимости смещения и разброса от параметра.
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes = axes.ravel()
    
    for idx, regressor in enumerate(regressors[:4]):
        biases = []
        variances = []
        
        for param_value in parameter_values:
            params = default_params.copy()
            params[parameter_name] = param_value
            
            if parameter_name == 'sample_size':
                bias, var = compute_bias_variance(
                    estimator=regressor,
                    sample_generator=generators[0],
                    sample_size=param_value,
                    n_samples=params['n_samples'],
                    n_objects=params['n_objects']
                )
            elif parameter_name == 'n_samples':
                bias, var = compute_bias_variance(
                    estimator=regressor,
                    sample_generator=generators[0],
                    n_samples=param_value,
                    sample_size=params['sample_size'],
                    n_objects=params['n_objects']
                )
            else:  # n_objects
                bias, var = compute_bias_variance(
                    estimator=regressor,
                    sample_generator=generators[0],
                    n_objects=param_value,
                    n_samples=params['n_samples'],
                    sample_size=params['sample_size']
                )
            
            biases.append(bias)
            variances.append(var)
        
        ax = axes[idx]
        ax.plot(parameter_values, biases, 'b-', label='Bias²', linewidth=2)
        ax.plot(parameter_values, variances, 'r-', label='Variance', linewidth=2)
        ax.set_xlabel(parameter_name)
        ax.set_ylabel('Value')
        ax.set_title(f'{regressor.__class__.__name__}')
        ax.legend()
        ax.grid(True)
    
    plt.tight_layout()
    plt.show()

default_params = {
    'sample_size': 200,
    'n_samples': 50,
    'n_objects': 1000
}

print("Зависимость от размера выборки (sample_size):")
plot_parameter_vs_bias_variance(
    parameter_name='sample_size',
    parameter_values=[50, 100, 200, 500, 1000],
    default_params=default_params
)

print("Зависимость от количества выборок (n_samples):")
plot_parameter_vs_bias_variance(
    parameter_name='n_samples',
    parameter_values=[10, 20, 50, 100, 200],
    default_params=default_params
)


print("Зависимость от количества тестовых объектов (n_objects):")
plot_parameter_vs_bias_variance(
    parameter_name='n_objects',
    parameter_values=[100, 500, 1000, 2000, 5000],
    default_params=default_params
)
```

sample_size: С увеличением размера выборки смещение уменьшается, особенно для сложных моделей. Разброс также уменьшается.

n_samples: С увеличением количества выборок оценки становятся более стабильными (дисперсия оценок уменьшается).

n_objects: С увеличением количества тестовых объектов оценки становятся более точными, но вычислительная сложность растет.
